```python
from google.colab import drive
import pandas as pd
import glob
import os

drive.mount('/content/drive')
```

```python
file_path ='/content/drive/MyDrive/IDX exchange/'
files = glob.glob(os.path.join(file_path, "*.csv"))

#----------- SOLD FILES-----------
#loads sold files
sold_files= sorted(glob.glob(os.path.join(file_path, 'CRMLSSold[0-9]*.csv')))

sold_dfs=[]

#row count before concatentation
print(f"Sold files row count: {len(sold_files)}")

for f in sold_files:
  df=pd.read_csv(f, low_memory=False)
  sold_dfs.append(df)

#concats sold files
sold = pd.concat(
    sold_dfs,
    ignore_index = True
)
#row count after concatentation
print(f"Combined sold files row count: {len(sold)}")


#-----------Listing Files-----------
#load listing files
listing_files= sorted(glob.glob(os.path.join(file_path, 'CRMLSListing[0-9]*.csv')))

listing_dfs=[]

#row count before concatentation
print(f"Listings files row count: {len(listing_files)}")

for f in listing_files:
  df=pd.read_csv(f, low_memory=False)
  listing_dfs.append(df)

#concats listing files
listings = pd.concat(
    listing_dfs,
    ignore_index = True
)
#row count after concatentation
print(f"Combined listing files row count: {len(listings)}")
```

```python
#Residential Only Filter

sold_before =len(sold)
listing_before = len(listings)

#residential filter
sold_residential = sold[
    listings["PropertyType"] == "Residential"
]
listing_residential = listings[
    listings['PropertyType'] == 'Residential'
]
print(f'Sold before filter: {sold_before}')
print(f'Sold after filter: {len(sold_residential)}')

print(f'Listings before filter: {listing_before}')
print(f'Listings after filter: {len(listing_residential)}')
```

```python
# Save Residential Sold dataset
sold_residential.to_csv(
    os.path.join(file_path, 'CRMLSSold_Residential.csv'),
    index=False
)

# Save Residential Listings dataset
listing_residential.to_csv(
    os.path.join(file_path,'CRMLSListing_Residential.csv'),
    index=False
)

print('Files saved successfully')
```
